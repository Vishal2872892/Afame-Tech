```python
data.drop(['EmployeeNumber', 'Age', 'Gender', 'MaritalStatus'], axis = 1, inplace = True)
```

```python
for attribute in data.columns:
  print(attribute)
  if data[attribute].dtype == object:
    print('Categorical. Values: {}\n'.format(data[attribute].unique()))
  else:
    print('Numeric. Range: {} to {}\n'.format(data[attribute].min(), data[attribute].max()))
```

```python
data.drop(['EmployeeCount', 'Over18', 'StandardHours'], axis = 1, inplace = True)
```

```python
data.head()
```

```python
sns.set_theme()
sns.countplot(data = data, x = 'Attrition', palette='crest')
```

```python
data['Attrition'] = data['Attrition'].apply(lambda val: 1 if val == 'Yes' else 0)
```

```python

data.corr()
sns.clustermap(data=data.corr(),vmin=-1,cmap='coolwarm')
```

```python
sns.barplot(data=data.corr()['Attrition'][1:].reset_index(), x='Attrition', y = 'index', palette = 'crest')
```

```python
sns.catplot(data=data,x='BusinessTravel',y='MonthlyIncome',hue='Attrition',order=['Non-Travel','Travel_Rarely','Travel_Frequently'],dodge=True,palette='crest')
```

```python
for attribute in data.columns:
    if data[attribute].dtype == object:
        print('{} : {}\n'.format(attribute, data[attribute].unique()))
```

```python
data['OverTime'] = data['OverTime'].apply(lambda val: 1 if val == 'Yes' else 0)
```

```python
def convert_travel(value):
    if value == 'Non-Travel':
        return 0
    elif value == 'Travel_Rarely':
        return 1
    else:
        return 2
```

```python
data['BusinessTravel'] = data['BusinessTravel'].apply(convert_travel)
```

```python
data_final = pd.get_dummies(data=data,columns=['Department','EducationField','JobRole'],drop_first=True)
```

```python
data_final.to_csv('Machine-Learning-Employee-Attrition.csv',index=False) #save dataset as csv for future use
data_final.describe()
```

```python
from sklearn.model_selection import train_test_split
X = data_final.drop(columns = 'Attrition')
y = data_final['Attrition']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=36)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(random_state=36) #default hyper-parameters, other than random state, to ensure consistency
random_forest.fit(X_train, y_train)
```

```python
random_forest_predict = random_forest.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score
```

```python
sns.heatmap(confusion_matrix(y_test, random_forest_predict),annot=True,cmap='crest',fmt='g')
plt.title('Confusion Matrix for Random Forest')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
print(classification_report(y_test,random_forest_predict))
```

```python

np.round(roc_auc_score(y_test,random_forest.predict_proba(X_test)[:, 1]),3)
```

```python

balanced_random_forest = RandomForestClassifier(class_weight='balanced',random_state=36)
balanced_random_forest.fit(X_train,y_train)
balanced_predict = balanced_random_forest.predict(X_test)
```

```python

sns.heatmap(confusion_matrix(y_test, balanced_predict),annot=True,cmap='crest',fmt='g')
plt.title('Confusion Matrix for Balanced Random Forest')
plt.xlabel('Predicted')

plt.ylabel('Actual')
```

```python
print(classification_report(y_test,balanced_predict))
```

```python
np.round(roc_auc_score(y_test,balanced_random_forest.predict_proba(X_test)[:, 1]),3)
```

```python
from sklearn.model_selection import GridSearchCV
parameter_grid = {'max_features' : ['sqrt','log2',0.3],'n_estimators' : [10,100,1000],'class_weight' : [{0:1,1:1},{0:1,1:10},{0:1,1:100},{0:10,1:1}], 'max_depth' : [8,16,32,None], 'random_state' : [36]}
```

```python
grid_search = GridSearchCV(RandomForestClassifier(), parameter_grid, n_jobs=-1, refit=True, verbose=0, scoring='roc_auc')
```

```python
grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_params_)
```

```python
grid_search_predict = grid_search.predict(X_test)
```

```python

sns.heatmap(confusion_matrix(y_test, grid_search_predict),annot=True,cmap='crest',fmt='g')
plt.title('Confusion Matrix for Grid Search Optimized Random Forest')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
print(classification_report(y_test,grid_search_predict))
```

```python

np.round(roc_auc_score(y_test,grid_search.predict_proba(X_test)[:, 1]),3)
```
