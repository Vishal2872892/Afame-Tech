```python
def man_wom_chi(passanger):
  age = passanger['Age']
  sex = passanger['Sex']

  return 'child' if age < 16 else sex
titanic_df['Person'] = titanic_df.apply(man_wom_chi, axis = 1)
```

# What deck were the passengers on and how does that relate to their class?

```python
dock_df = titanic_df.dropna(axis = 0)
dock_df.head()
```

```python
#How do we find out what deck a passenger was assigned?
# The Cabin attribtute holds that data. Intuitively, the cabin number of a passenger is a
#combination of the deck they're on, and their room number.
# So a passenger on deck 'C' will have a cabin number in CXXX format, where XXX can be a room number.
#We just need to create a python method to extract first character from the cabin information.
#Grabbing the deck from the cabin numbers
def get_level(passanger):
  cabin = passanger['Cabin']
  return cabin[0]
```

```python
dock_df['level'] = dock_df.apply(get_level, axis = 1)
```

```python
dock_df.head()
```

```python
 #Everything seems to work, so now we can check out how many passengers belonged to different decks.
sns.catplot(x='level', data = dock_df, palette = 'winter_d', kind = 'count')
```

```python
#It feels like passengers belonging to the upper classes are indeed assigned to the upper decks.
# Decks A,B and C were assigned only to first class passengers,
# and we start seeing third class passengers only from level E.
# Though, we have to keep in mind that cabin information was missing for more than 3/4 of our passengers.
sns.catplot(x = 'level', data = dock_df, hue ='Pclass', kind = 'count')
```

#Where did the passengers come from?

```python
 # The Embarked attribute contains data for
# the passengers' port of embarkation (C = Cherbourg; Q = Queenstown; S = Southampton).
sns.catplot(data = dock_df, x = 'Embarked', hue = 'Pclass', kind = 'count', color = 'red')
```

# Who was with their family?


```python
titanic_df.head()
```

```python
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
```

```python
titanic_df.tail()
```

```python
titanic_df['Alone'].loc[titanic_df['Alone']>0] = 'No'
titanic_df['Alone'].loc[titanic_df['Alone']==0] = 'Yes'
```

```python
titanic_df.head()
```

```python
sns.catplot(data = titanic_df, x = 'Alone', kind = 'count', color = 'orange')
# The information is pretty simple to digest now. The majority of people on the ship had family on the ship.
# But did that make a difference in their chances of survival?
```

```python
# What factors helped someone survive the sinking?
```

```python
sns.catplot(data = titanic_df, x= 'Pclass', y = 'Survived', kind = 'point')
# To not much surprise, it seems like lower class passengers had a lower chance of survival.
# But wait, we saw earlier that there were more men than woman or children, in the
# 3rd class than in any other class. From the movie 'Titanic', you might remember the
# 'Women and children first!' protocol that the ship's administration followed while
# putting passengers on the lifeboats
```

```python
sns.catplot(data = titanic_df, x = 'Pclass', y = 'Survived', hue = 'Person', kind = 'point')
# To not much surprise, it seems like lower class passengers had a lower chance of survival.
# But wait, we saw earlier that there were more men than woman or children, in the
# 3rd class than in any other class. From the movie 'Titanic', you might remember the
# 'Women and children first!' protocol that the ship's administration followed while
# putting passengers on the lifeboats.
```

```python
# Let's see if this is really the case, by using the 'Person' column as the hue in the factorplot.
sns.lmplot(data = titanic_df, x = 'Age', y = 'Survived')
```

```python
 # Let's see if there's a relation between the age and survival.

sns.lmplot(data = titanic_df, x = 'Age', y = 'Survived', hue = 'Pclass')
```

```python
# The general concensus from the above two figures seems to be that older people were less likely to survive.
sns.catplot(x='Survived',y='Alone',data=titanic_df)
```
